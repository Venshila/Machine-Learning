```python
import pandas as pd 
# alltypes of table formats are listed under pandas
```

```python
dataset=pd.read_csv("Salary_Data.csv")
# loading the data set file
```

```python
dataset
# printing the dataset
```

```python
independent=dataset[["YearsExperience"]]
independent

independent means input, from the dataset we are giving yearsExperience as input
```

```python
dependent=dataset[["Salary"]]
dependent

# dependent means output, salary column is printed
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)

# sklearn is the library function and importing train_test_split as the class
#X-train,X_test-- X stands as the input for train and test method,t_train,y_test -- y as the output
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)

# from the sklearn library we are importing LinearRegression class as like oops method, assigning the regressor as variable, by calling the fit function
# X_train weight is calculated predefinedly with coef_ and the bais y-test is calculated by intercept_
#regressor is the variable
```

```python
weight=regressor.coef_
```

```python
weight
```

```python
bais=regressor.intercept_
```

```python
bais
```

```python
y_pred=regressor.predict(X_test)

# y-test which means y-pred is getiing calculated by calling the .predict function
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)

# evaluating metrics is calculated as actual value-predicted value, assigning to the variable r-score, r2-score class is imported
```

```python
r_score
```

```python
import pickle
filename="finalized_model_linear.sav"

# to save the model pickle library function is used here file name we are getting as the finaized-model-linear.sav
# we are creating the variable filename and naming it as "finalized_model_linear.sav"
```

```python
pickle.dump(regressor,open(filename,'wb'))

#  here opening the file is happend and wb denotes write binary
#regressor stores all the functions
#so we are opening the corresponding file and dumping it
```

```python
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
result=loaded_model.predict([[13]])

# dumped file is loaded from the saved file
# as the result loaded model is assigned with pickle.load function to open the finalized file
```
